```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/diabetes_prediction_dataset_modified.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
continous_columns = ["age", "bmi", "HbA1c_level", "blood_glucose_level"]
```

```python
categorical_columns = ["gender", "hypertension", "heart_disease", "smoking_history", "diabetes"]
```

```python
for i, column in enumerate(continous_columns, 1):
    plt.rcParams["figure.figsize"] = (20, 20)
    plt.subplot(2,2, i)
    plt.title(column)
    sns.histplot(data=df, x=column, kde=True)
```

```python
for i, column in enumerate(categorical_columns, 1):
    plt.rcParams["figure.figsize"] = (20, 20)
    plt.subplot(3,2, i)
    plt.title(column)
    df[column].value_counts().plot(kind="bar")
    
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df["blood_glucose_level"] = df["blood_glucose_level"].fillna(df["blood_glucose_level"].mean())
```

```python
df["smoking_history"] = df["smoking_history"].fillna("never")
```

```python
df.info()
```

```python
for i, column in enumerate(continous_columns, 1):
    plt.rcParams["figure.figsize"] = (20, 20)
    plt.subplot(2,2, i)
    plt.title(column)
    sns.boxplot(data = df, y = column)
```
